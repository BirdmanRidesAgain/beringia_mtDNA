## Introduction

We need to be able to find genes in these mitogenomes/mitochondrial fragments.
To do this, we are using an annotator called [mitoZ](https://github.com/linzhi2013/MitoZ) ([Meng et al.](https://academic.oup.com/nar/article/47/11/e63/5377471?login=true)).

We'll run this on all the fasta files that came out of NOVOPlasty.

## Methods

See [code documentation](#code-documentation) for software implementation details.

### Annotating different levels of completeness

Because NOVOPlasty sometimes fails to fully assemble (and produces fragmented assemblies), we ran the annotator with the following level of priority:

| Tier | Name | Description | Number |
| - | - | - | - |
| 1 | `Circularized_assembly_*.fasta`| Unambiguously resolved mitogenome | 147 |
| 2 | `Option_1_*.fasta` | Plausible assembly, but more than one ordering present | 19 `Option_1`, 3 `Option_1` + `Option_2` |
| 3 | `Contigs_*.fasta` | One or more contigs present | 150 |
| 4 | `contigs_tmp_*.txt` | Individual, small contigs. Most have file sizes of 0. Not likely to yield successful annotation. | 93 |

## Code documentation

We run `mitoz` through the conda environment `mitoz-x64`.
On an OSX machine, you install it with the conda environment in [mitoz-x64.yml](resources/mitoz-x64.yml).

We then ran the program using [annotate_mitogenome.py](scripts/annotate_mitogenome.py) to call `mitoz`, and [batch_annotate_assemblies.py](scripts/batch_annotate_assemblies.py) to iterate across all files in the directory.

```bash
# sample code for a single run
conda activate mitoz-x64 
python scripts/annotate_mitogenome.py -i $FA -s sample1 -v
```

Outputs can be found at `mitogenome_assemblies`.
A log file can be found at [batch_annotate_assemblies.log](mitogenome_annotations/batch_annotate_assemblies.log).
